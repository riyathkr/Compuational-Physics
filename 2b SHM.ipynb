```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Euler and Runge Kutta solver functions:

```python
def OdeEuler(x0,t0,v0,tf,h):
    t=np.arange(t0,tf,h);
    x=np.zeros(len(t))
    v=np.zeros(len(t))
    x[0]=x0
    v[0]=v0
    for n in range(len(t)-1):
        x[n+1]=x[n]+h*v[n]
        v[n+1]=v[n]-h*x[n]
    return(t,x,v)
```

```python
def RungeKutta2(x0,t0,v0,tf,h):
    t=np.arange(t0,tf,h)
    x=np.zeros(len(t))
    v=np.zeros(len(t))
    x[0]=x0
    v[0]=v0
    for n in range(len(t)-1):
        k1x=v[n]
        k1v=-x[n]
        k2x=v[n]+h*k1v
        k2v=-x[n]-h*k1x
        
        x[n+1] = x[n] + (h/2)*(k1x+k2x);
        v[n+1] = v[n] + (h/2)*(k1v+k2v);
        
    return(t,x,v)
```

```python
def RungeKutta4(x0,t0,v0,tf,h):
    t=np.arange(t0,tf,h)
    x=np.zeros(len(t))
    v=np.zeros(len(t))
    x[0]=x0
    v[0]=v0
    
    for n in range(len(t)-1):
        k1x = v[n]
        k1v = -x[n]
        
        k2x = v[n] + (h/2)*k1v
        k2v = -x[n] - (h/2)*k1x
        
        k3x = v[n] + (h/2)*k2v
        k3v = -x[n] - (h/2)*k2x
        
        k4x = v[n] + h*k3v
        k4v = -x[n] - h*k3x
        
        x[n+1] = x[n] + (h/6)*(k1x+2*k2x+2*k3x+k4x);
        v[n+1] = v[n] + (h/6)*(k1v+2*k2v+2*k3v+k4v);
    return(t,x,v)
        
    
```

# Plotting the solution obtained through Euler and Runge Kutta methods:

# Euler method:

```python
t1,x1,v1=OdeEuler(1,0,0,10,0.1)
t2,x2,v2=RungeKutta2(1,0,0,10,0.1)
t3,x3,v3=RungeKutta4(1,0,0,10,0.1)
plt.plot(t1,x1,label="x(t)")
plt.plot(t1,v1,label="v(t)")
plt.xlabel("time ->")
plt.ylabel("")
plt.title("Solved with Euler's method")
plt.grid()
plt.legend()
plt.show()
```

# Analysis: 

Amplitude of x(t) and v(t) increases with time -> Energy of the oscillating particle increases with time for the solution obtained through Euler's method. 
The error propogates, thereby increasing the error associated with x and v with time.
The phase difference between x(t) and v(t) doesn't remain constant.


# Runge Kutta 2nd order

```python
plt.plot(t2,x2,label="x(t)")
plt.plot(t2,v2,label="v(t)")
plt.xlabel("time ->")
plt.ylabel("")
plt.title("Solved with Runge Kutta 2nd order method")
plt.grid()
plt.legend()
plt.show()
```

# Analysis : 
-Amplitude of x(t) and v(t) is constant

-Energy associated with the mass is constant

-The phase difference between the two is constant

-The solution is more accurate compared to that obtained through Euler's method
